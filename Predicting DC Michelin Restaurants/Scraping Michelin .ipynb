```python
#pulling data from Michelin - first, one-star restaurants

import urllib
from bs4 import BeautifulSoup
onestar = 'https://www.viamichelin.com/web/Restaurants/Restaurants-United_States?stars=1'
html = urllib.urlopen(onestar).read()
soup = BeautifulSoup(html,"html.parser")
```

```python
print soup.prettify()
```

```python
def getrest(soup):
    for row in soup.findAll('div',{'class':'poi-item-info'}):
        rest_list.append((row.find('div',{'class':'poi-item-name truncate'}).get_text()).strip('\n'))
        address_list.append((row.find('div',{'class':'poi-item-address truncate'}).renderContents()).strip('\n').strip('\t').strip('\n'))
```

```python
rest_list = []
address_list = []
```

```python
for i in range(1,6):
    url = 'https://www.viamichelin.com/web/Restaurants/Restaurants-United_States?stars=1&page=%d' %(i)
    html = urllib.urlopen(url).read()
    soup = BeautifulSoup(html,"html.parser")
    getrest(soup)
```

```python
for stars in range(2,4):
    url = 'https://www.viamichelin.com/web/Restaurants/Restaurants-United_States?stars=%d'%(stars)
    html = urllib.urlopen(url).read()
    soup = BeautifulSoup(html,"html.parser")
    rest_list = []
    address_list = []
    getrest(soup)
    stars_add = pd.DataFrame({'restaurant' : rest_list, 'address':address_list})
    stars_add['stars'] = stars
    michelin = michelin.append(stars_add)
    michelin = michelin.reset_index()
    del michelin['index']
```
